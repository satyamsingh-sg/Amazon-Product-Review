```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
```

```python
df = pd.read_csv("amazon.csv")
data = pd.DataFrame(df, columns = ['Asin','country'])
print(data)
```

```python


def get_title(productsoup):
     
    try:
        title = productsoup.find("span", attrs={"id":'productTitle'})
 
        
        title_value = title.string
 
        title_string = title_value.strip()
 
 
    except AttributeError:
        title_string = "NULL"   
 
    return title_string

def get_price(productsoup):
 
    try:
        price = productsoup.find("span", attrs={'class':'a-price-whole'}).string.strip()
 
    except AttributeError:
        price = "NULL"  
 
    return price
 
# Function to extract Product Rating
def get_rating(productsoup):
 
    try:
        rating = productsoup.find("i", attrs={'class':'a-icon a-icon-star a-star-4-5'}).string.strip()
         
    except AttributeError:
         
        try:
            rating = productsoup.find("span", attrs={'class':'a-icon-alt'}).string.strip()
        except:
            rating = "NULL" 
 
    return rating

def reviws_product(productsoup):
    try:
        review_count = productsoup.find("span", attrs={'id':'acrCustomerReviewText'}).string.strip()
         
    except AttributeError:
        review_count = "NULL"   
 
    return review_count
 
# Function to extract Availability Status
def get_availability(productsoup):
    try:
        available = productsoup.find("div", attrs={'id':'availability'})
        available = available.find("span").string.strip()
 
    except AttributeError:
        available = "NULL"  
 
    return available 

def image_fun(productsoup):
    try:
        available = productsoup.find("img", attrs={"id":"landingImage"})

        available = available.find("span").string.strip()
 
    except AttributeError:
        available = "NULL"  
 
    return available 

if __name__ == '__main__':
 
    # Headers for request
    HEADERS = ({'User-Agent':
                'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
                'Accept-Language': 'en-US, en;q=0.5'})
 
    # The webpage URL
    arr=[]
    for i in range(len(data)):
      URL = "https://www.amazon."+str(df.loc[i, "country"])+"/dp/"+str(df.loc[i, "Asin"])+"/"
  
      # HTTP Request
      webpage = requests.get(URL, headers=HEADERS)
      

      
      productsoup = BeautifulSoup(webpage.content, "lxml")
      
      title=get_title(productsoup)
     
      img=image_fun(productsoup)
     
      product_title=get_title(productsoup)
    
      price=get_price(productsoup)
      
      rating=reviws_product(productsoup)
     
      avail=get_availability(productsoup)
      dict={"imageUrl":img,"productTitle":product_title,"price":price,"rating":rating,"available":avail}
      arr.append(dict)
    
```

```python
print(arr)
```

```python
with open("sample.json", "w") as outfile:
      json.dump(arr, outfile)
```
